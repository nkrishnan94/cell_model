```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.animation import PillowWriter



fig = plt.figure(figsize=(20,12))
ax = fig.add_subplot(111, projection='3d')

ims = []
f=files1[-1]

for f in files1[:20]:
    
    arr = np.loadtxt(f,delimiter=", ")


    n = 100

    # For each set of style and range settings, plot n random points in the box
    # defined by x in [23, 32], y in [0, 100], z in [zlow, zhigh].
    #for m in arr:
    #    ax.scatter(*m[1:4],s= (m[4])**2,cmap='viridis')
    x=arr[:,1].flatten()
    y= arr[:,2].flatten()
    z=arr[:,3].flatten()

    ax.set_xlabel('X Label')
    ax.set_ylabel('Y Label')
    ax.set_zlabel('Z Label')
    ax.set_xlim([0,90])
    ax.set_ylim([0,1440])

    ax.set_zlim([0,400])


    x=x[z>0]
    y=y[z>0]
    c=arr[:,5][z>0]*10**6

    ax.set_zlim([0,700])
    c = (np.sum(arr[:,5:]**2,axis=1)**.5)[z>0]
    z= z[z>0]
    im = ax.scatter(x, y, z,c=c, s=40,vmin=0,vmax=np.max(np.sum(arr[:,5:]**2,axis=1)**.5),cmap='viridis' );
    plt.colorbar(im)
    ims.append([im])
    ax.view_init(0, 360)
    print(f)
    plt.show()





ani = animation.ArtistAnimation(fig, ims, interval=20000, blit=True,
                                repeat_delay=500)

writer = PillowWriter(fps=20)
ani.save("two_slit.gif", writer=writer)

plt.show()
```
