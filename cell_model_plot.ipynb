```python
import numpy as np
import glob
import os
import matplotlib.pyplot as plt



from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import


import matplotlib.pyplot as plt
import numpy as np
```

```python
files=glob.glob("CPD/prof_T*.txt")
```

```python
nums=[]
for f in files:
    nums.append(int(f.split("T_")[1].split("_")[0]))
    
files =np.array(files)[np.argsort(nums)]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ims = []
f=files[-1]

arr = np.loadtxt(f,delimiter=", ")


n = 100

# For each set of style and range settings, plot n random points in the box
# defined by x in [23, 32], y in [0, 100], z in [zlow, zhigh].
#for m in arr:
#    ax.scatter(*m[1:4],s= (m[4])**2,cmap='viridis')
x=arr[:,1].flatten()
y= arr[:,2].flatten()
z=arr[:,3].flatten()

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_xlim([0,90])
ax.set_ylim([0,800])

ax.set_zlim([0,400])

x=x[z>0]
y=y[z>0]
c=arr[:,5][z>0]
z= z[z>0]
ax.set_zlim([0,700])
im = ax.scatter(x, y, z,c=c, s=4,
                vmin=np.min(c),vmax=np.max(c),cmap='viridis' );
plt.colorbar(im)
ax.view_init(0, 360)
print(f)
plt.show()
```
